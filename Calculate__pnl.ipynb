```python
import numpy as np

#store data values
buy_price = np.array((100,200,300,500))
sell_price = np.array((120,200,290,505))
quantity = np.array((10000,3000,5000,156000))
Pnl = (sell_price - buy_price) * quantity
print ("profit", Pnl)
# Print data types
print (type(buy_price))
print (type(Pnl))
```
