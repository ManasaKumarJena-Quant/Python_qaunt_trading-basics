```python
import numpy as np
#input buy,sell,qty and calculate pnl where price is >500
buy_price = np.array((100,200,310,410,500,600,800,900,550))
sell_price = np.array((1000,300,1500,1000,1500,1600,500,1800,7000))
quantity = np.array((356,234,876,653,897,567,987,435,947))
Pnl = (sell_price - buy_price) * quantity
#condition is price is greator than >500
print ("print Pnl for all trades",Pnl)
print ("print Pnl profit is > 500")
print (Pnl[Pnl > 500])
```
