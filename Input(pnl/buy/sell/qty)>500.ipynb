```python

import numpy as np

#Write program: Input buy, sell, quantity â†’ print PnL and message if profit > 500.

# Input values
buy = float(input("Enter buy price: "))
sell = float(input("Enter sell price: "))
quantity = int(input("Enter quantity: "))

# Calculate PnL
pnl = (sell - buy) * quantity

# Print PnL
print("PnL:", pnl)

# Check if profit > 500
if pnl > 500:
    print("Great! Profit is more than 500.")
else:
    print("watch the markets")
```
